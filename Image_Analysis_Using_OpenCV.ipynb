# Detecting Faces

```python
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
imag=cv2.imread('group5.jpg')
```

```python
image=cv2.resize(imag,(3000,1800))
```

```python
img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Resize
image=cv2.resize(image,(600,900))
```

```python
# Convert to gray scale image

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
faces=face_cascade.detectMultiScale(gray)
```

```python
len(faces)
```

```python
# Display the faces in the image

for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2.imshow('Faces',image)
cv2.waitKey(0)
```

```python
image=cv2.imread('group1.jpeg')
```

```python
#Lets first build a matrix of same shape as image with values of 100
I = np.ones(image.shape,dtype="uint8") * 50 #matrix with 100 as values with the same shape as image
#Open CV

cv2_add = cv2.add(image, I)
cv2_sub = cv2.subtract(image, I)
#Numpy
np_add = np.add(image, I)
np_sub = np.subtract(image, I)

fig,ax = plt.subplots(2,2,figsize=(15,10))
ax[0,0].imshow(cv2_add)
ax[0,0].set_title("CV add 100")
ax[0,1].imshow(cv2_sub)
ax[0,1].set_title("CV sub 100")
ax[1,0].imshow(np_add)
ax[1,0].set_title("Numpy add 100")
ax[1,1].imshow(np_sub)
ax[1,1].set_title("Numpy sub 100")
```

# Drawing

```python
# (0,0,0) denotes black - board with black color

board = np.zeros((255,255,3),dtype='uint8')
plot_BGR_to_RGB(board,plt)
plt.axis('off')
```

```python
red=(255,0,0)
green=(0,255,0)
blue=(0,0,255)

cv2.line(board,(0,255),(255,0),green,5)
cv2.line(board,(0,0),(500,500),red)

cv2.rectangle(board,(10,10),(200,200),red,3) # Draw Rectangle
cv2.rectangle(board,(200,200),(300,300),blue,-1)

(centerX,centerY)=(board.shape[1]//2,board.shape[0]//2)

cv2.circle(board,(centerX,centerY),70,green,3) # Draw Circle`
plot_BGR_to_RGB(board,plt)
```

```python
# (0,0,0) denotes black

board_nxt=np.zeros((400,400,3),dtype='uint8')

# Some random drawings:
max_size=400

for i in range(0,50):
    r = np.random.randint(1,110)
    color = np.random.randint(0,256,size=3).tolist()
    pt=np.random.randint(0,max_size,size=2)
    end_pt = np.random.randint(0,max_size,size=2)
    
    cv2.circle(board_nxt,tuple(pt),r,color,-1)
    
plot_BGR_to_RGB(board_nxt,plt)
```

```python
board_nxt1=np.zeros((400,400,3),dtype='uint8')+255

width = 5
val = 0

for i in range(0,400,5):
    for j in range(0,400,5):
        spt = (i,j)
        ept = (i+width , j+width)
        color = (0,0,0) if val % 2 else (255,255,255)
        cv2.rectangle(board_nxt1,spt,ept,color,-1)
        val=val+1
    val=val+1
    
centerx,centery = board_nxt1.shape[1] // 3, board_nxt1.shape[0] // 3
centera,centerb = board_nxt1.shape[1] // 2, board_nxt1.shape[0] // 2
centerc,centerd = board_nxt1.shape[1] // 5, board_nxt1.shape[0] // 5

cv2.circle(board_nxt1, (centerx,centery), 20, (255,0,0) , -1)
cv2.circle(board_nxt1, (centera,centerb), 70, (0,255,0) , -1)
cv2.circle(board_nxt1, (centerc,centerd), 50, (0,0,255) , 6)

plot_BGR_to_RGB(board_nxt1,plt)
```
